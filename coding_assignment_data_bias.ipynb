Transforming the excel spreadsheet with the sample comments, expected value, perspective score, and perspective predicted value into a dataframe.

```python
import pandas as pd
df = pd.read_csv("coding_assignment_data_bias.csv")
df
```

Results:

Class-wise accuracy for toxic and non-toxic comments:

```python
toxic_accuracy = sum((df['Expected Value'] == 'Toxic') & (df['Predicted Value'] == 'Toxic')) / sum(df['Expected Value'] == 'Toxic')
non_toxic_accuracy = sum((df['Expected Value'] == 'Non-Toxic') & (df['Predicted Value'] == 'Non-Toxic')) / sum(df['Expected Value'] == 'Non-Toxic')

print(f"The accuracy for toxic comments is {toxic_accuracy:.2%}")
print(f"The accuracy for non toxic comments is {non_toxic_accuracy:.2%}")
```

Religion-wise accuracy for comments:

```python
islam_accuracy = sum((df['Predicted Value'][:10] == df['Expected Value'][:10])) / len(df['Predicted Value'][:10])
buddhism_accuracy = sum((df['Predicted Value'][10:] == df['Expected Value'][10:])) / len(df['Predicted Value'][10:])

print(f"The accuracy for Islam-themed comments is {islam_accuracy:.2%}")
print(f"The accuracy for Buddhism-themed comments is {buddhism_accuracy:.2%}")
```
